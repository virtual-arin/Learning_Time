```python
#Creating Data
data = {
    "Name": ["Arin", "Rishita", "Vaishnavi", "Advik", "Vishal", "Radhika"],
    "Grade": [65, 99, 84, 68, 86, 88],
    "State": ["Uttar Pradesh", "Gujarat", "Haryana","Uttarakhand", "Uttarakhand", "Madhya Pradesh"],
    "Gender": ["Male", "Female", "Female", "Male","Male", "Female"],
    "Salary": [60000, 950000, 40000, 20000, 10000, 65000]
}
df = pd.DataFrame(data)
df
```

```python
df.describe() # generate descriptive statistics of a dataFrame 
```

#### Filtering based on condition

```python
salary = df[df["Salary"]> 50000]
salary
```

```python
lower_salary = df[(df["Salary"] > 10000) & (df["Salary"] < 50000)]
lower_salary
```

```python
df["Name"] #returns single column
```

```python
df[["Name", "Salary"]] #returns multiple columns
```

```python
higher_rankers = df[(df["Grade"] > 85) & (df["Grade"] < 90)]
higher_rankers 
```

```python
passed = df[(df["Grade"] > 50) & (df["Grade"] < 70)]
passed
```
