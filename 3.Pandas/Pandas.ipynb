## Handling Missing data

```python
#Creating Data
data = {
    "Name": ["Arin", "Rishita", "Vaishnavi", "Advik", "Vishal", "Radhika"],
    "Grade": [None, 99, 84, 68, 86, 88],
    "State": ["Uttar Pradesh", None, "Haryana","Uttarakhand", "Uttarakhand", "Madhya Pradesh"],
    "Gender": ["Male", "Female", None, "Male","Male", "Female"],
    "Salary": [60000, 950000, 40000, None, 10000, 65000]
}
df = pd.DataFrame(data)
df
```

```python
df.isnull() #return true if value is missing
```

```python
df.isnull().sum() #return count of value if it is missing
```

```python
#update missing with zero
df.fillna(0, inplace=True)
df
```

```python
#Remove if data is missing
df.dropna(axis=1, inplace=True)
df
```

## Sorting & aggregation

```python
#Creating Data
data = {
    "Name": ["Arin", "Rishita", "Vaishnavi", "Advik", "Vishal", "Radhika"],
    "Grade": [65, 99, 84, 68, 86, 88],
    "State": ["Uttar Pradesh", "Gujarat", "Haryana","Uttarakhand", "Uttarakhand", "Madhya Pradesh"],
    "Gender": ["Male", "Female", "Female", "Male","Male", "Female"],
    "Salary": [60000, 950000, 40000, 20000, 10000, 65000]
}
df = pd.DataFrame(data)
df
```

```python
#sorted in descending based on grade
df.sort_values(by="Grade", ascending=False, inplace=True)
df
```

```python
#sorted in ascending based on grade
df.sort_values(by="Grade", ascending=True, inplace=True)
df
```

```python
average_salary = df['Salary'].mean()
average_salary
```

```python
grouped = df.groupby("Grade")['Salary'].sum()
grouped
```

```python
grouped = df.groupby(["Grade", "Name"])['Salary'].sum()
grouped
```

## Merging and Concatenation

```python
customers = pd.DataFrame({
    "CustomerID": [1, 2, 3, 4],
    "Name": ["Ram", "Raju", "Mohan", "Shiv"]
})
```

```python
orders = pd.DataFrame({
    "CustomerID":[1, 2, 3, 4],
    "Amount": [250, 500, 750, 100]
})
```

```python
merged_df = pd.merge(customers, orders, on="CustomerID", how="inner")
merged_df
```

```python
merged_df = pd.merge(customers, orders, on="CustomerID", how="left")
merged_df
```

```python
customers_df1 = pd.DataFrame({
    "CustomerID": [1, 2, 3],
    "Name": ["Ram", "Narayan", "Shiv"]
})
```

```python
customers_df2 = pd.DataFrame({
    "CustomerID": [5, 6, 7],
    "Name": ["Sita", "Lakshmi", "Parvati"]
})
```

```python
#vertically
concatenate = pd.concat([customers_df1, customers_df2], ignore_index=True)
concatenate
```

```python
#horizontally
concatenate = pd.concat([customers_df1, customers_df2], axis=1, ignore_index=True)
concatenate
```
