```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
X
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
len(dataset)
```

```python
len(X_test)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Classical ML

from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
pred = clf.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)

print(cm)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import accuracy_score

score = accuracy_score(y_test,pred)
print(score*100)
```

### Accuracy from Decision Tree Classifier: 80%

```python
## Random Forest

#Classical ML

from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier(n_estimators=200)
clf2.fit(X_train, y_train)
pred2 = clf2.predict(X_test)
```

# ANN - Let's get started!!

```python
# Now let's make the ANN!

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()
#units: It's an art, comes by experience, input+output/2
#initializer: how your weights are updated
#relu activation function
#input_dim: input shape (shape of the data, how many features do we have in the data)
# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))

# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the third  hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
len(X_train)
```

```python
# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
y_train[0]
# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 50)
```

```python
# Part 3 - Making predictions and evaluating the model
# Predicting the Test set results
y_pred_ann = classifier.predict(X_test)
y_pred_ann = (y_pred_ann > 0.5)

cm3 = confusion_matrix(y_test, y_pred_ann)
print(cm3)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import accuracy_score

score3 = accuracy_score(y_test,y_pred_ann)
print(score3*100)
```

```python
#Improvising the ANN - Hyper parameter optimization

from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from keras.models import Sequential
from keras.layers import Dense


def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 10))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier


classifier = KerasClassifier(build_fn = build_classifier)

parameters = {'batch_size': [10, 20],
              'epochs': [5, 6]}

              # Total combinations: 4

grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', n_jobs = -1, cv = 5)
grid_search = grid_search.fit(X_train, y_train)
grid_search.best_params_
grid_search.best_score_
```

```python
import numpy as np
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# 1. Define the model building function with default arguments
# We use meta to dynamically grab the number of features from X_train
def build_classifier(meta, loss='binary_crossentropy', optimizer='adam'):
    n_features_in_ = meta["n_features_in_"]

    classifier = Sequential()

    # First Hidden Layer - Dynamic input_dim based on your data
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim=n_features_in_))

    # Second Hidden Layer
    classifier.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))

    # Output Layer (Sigmoid for binary classification)
    classifier.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))

    classifier.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    return classifier

# 2. Initialize the SciKeras wrapper
# SciKeras uses 'model=' instead of 'build_fn='
classifier = KerasClassifier(model=build_classifier)

# 3. Define the Hyperparameter Grid
# Note: Use 'model__' prefix for arguments passed to build_classifier
parameters = {
    'batch_size': [10, 25],
    'epochs': [50, 100],
    'optimizer': ['adam', 'rmsprop'], # SciKeras can route 'optimizer' directly
    'loss': ['binary_crossentropy']   # categorical_crossentropy would fail here
}

# 4. Setup Grid Search
# n_jobs=-1 uses all CPU cores; cv=10 is standard cross-validation
grid_search = GridSearchCV(estimator=classifier,
                           param_grid=parameters,
                           scoring='accuracy',
                           cv=10,
                           n_jobs=-1)

# 5. Fit the model
grid_search = grid_search.fit(X_train, y_train)

# 6. Output Results
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_

print(f"Best Parameters: {best_parameters}")
print(f"Best Accuracy: {best_accuracy:.4f}")
```
