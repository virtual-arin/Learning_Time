```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import plot_tree
```

```python
x, y = make_classification(n_samples=200, n_features=5, n_redundant=0, n_informative=5, n_clusters_per_class=1)
```

```python
df = pd.DataFrame(x, columns=['Column1', 'Column2', 'Column3', 'Column4', 'Column5'])
df['target'] = y
```

```python
df.head()
```

```python
len(df)
```

```python
df.shape
```

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5], df.iloc[:, -1])
```

```python
plt.figure(figsize=(10,10))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rfc = RandomForestClassifier(max_features=2)
rfc.fit(df.iloc[:,:5], df.iloc[:, -1])
```

```python
plt.figure(figsize=(10,10))
plot_tree(rfc.estimators_[0])
plt.show()
```
