```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x, y = make_classification(n_samples=10000, n_features=10, n_informative=3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(x_train, y_train)
y_prediction_decision_tree_model = decision_tree_model.predict(x_test)
print('Decision Tree Accuracy: ', round(accuracy_score(y_test, y_prediction_decision_tree_model)*100, 2), '%')
```

# Bagging Classifier

```python
bag = BaggingClassifier(estimator=DecisionTreeClassifier(), 
                        n_estimators=500,
                        max_samples=0.5,
                        bootstrap=True,
                        random_state=42)
```

```python
bag.fit(x_train, y_train)
```

```python
bag_y_prediction = bag.predict(x_test)
```

```python
print('Bag Classifier Accuracy: ', round(accuracy_score(y_test, bag_y_prediction)*100, 2), '%')
```

# Random Forest Classifier

```python
random_forest_model = RandomForestClassifier(random_state=42, n_estimators=500)
random_forest_model.fit(x_train, y_train)
y_prediction_random_forest_model = random_forest_model.predict(x_test)
print('Random Forest Accuracy: ', round(accuracy_score(y_test, y_prediction_random_forest_model)*100, 2), '%')
```

# Bagging using SVM

```python
bag_svm = BaggingClassifier(estimator=SVC(), 
                        n_estimators=500,
                        max_samples=0.25,
                        bootstrap=True,
                        random_state=42)
```

```python
bag_svm.fit(x_train, y_train)
```

```python
y_prediction_bag_svm = bag_svm.predict(x_test)
print('Bag SVM Accuracy: ', round(accuracy_score(y_test, y_prediction_bag_svm)*100, 2), '%')
```

# Pasting

```python
pasting = BaggingClassifier(estimator=DecisionTreeClassifier(), 
                        n_estimators=500,
                        max_samples=0.25,
                        bootstrap=False,
                        random_state=42)
```

```python
pasting.fit(x_train, y_train)
y_prediction_pasting = pasting.predict(x_test)
print('Pasting Accuracy: ', round(accuracy_score(y_test, y_prediction_pasting)*100, 2), '%')
```
