```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
np.set_printoptions(suppress=True, precision=2)
x = df.iloc[:,0].values
y = df.iloc[:,-1].values
```

```python
x,y
```

```python
# train test split function
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train = x_train.reshape(-1, 1)
x_test = x_test.reshape(-1, 1)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
standard_scaling = StandardScaler()
x_train = standard_scaling.fit_transform(x_train)
x_test = standard_scaling.transform(x_test)
```

```python
x_test
```

```python
# Create Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Prediction
y_prediction = regressor.predict(x_test)
```

```python
plt.plot(y_test, color='red', label='Test')
plt.plot(y_prediction, color='green', label='Predictions')
plt.show()
```

# Regression Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_prediction, y_test)
print(mae)
```

```python
mse = mean_squared_error(y_prediction, y_test)
print(mse)
```

```python
print(np.sqrt(mse))
```

```python
r2 = r2_score(y_prediction, y_test)
print(r2)
```

```python
# Adjusted-R-Squared
n = x_test.shape[0]
k = x_test.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/(n-1-k)
print(adjusted_r2)
```
