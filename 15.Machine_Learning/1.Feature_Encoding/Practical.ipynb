```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.info()
```

```python
df.drop(columns=['CustomerId', 'RowNumber', 'Surname'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Gender.value_counts()
```

```python
# Handling missing values
df.Gender.mode()
```

```python
df['Gender'] = df['Gender'].fillna('Male')
```

```python
df.Gender.value_counts()
```

```python
df.info()
```

```python
# Label Encoding
label_encoding = preprocessing.LabelEncoder()
df['Gender_label'] = label_encoding.fit_transform(df.Gender.values)
```

```python
df.head()
```

```python
df.Gender_label.value_counts()
```

```python
# One hot encoding
one_hot_encoding = pd.get_dummies(df['Geography'], dtype=int)
one_hot_encoding
```
