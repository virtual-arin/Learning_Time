# Matplotlib
* Matplotlib is a comprehensive plotting library in Python, widely used for creating static, animated, and interactive visualizations. It provides a robust and flexible framework for generating a wide variety of plots, ranging from simple line plots to complex 3D visualizations.

```python
#Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title('Sine wave form')

# Plot the points using matplotlib
plt.plot(x, y)
plt.show()
```

```python
#Subplot()
#Compute the x & y coordinates for points on sine and cosine curves
x = np.arange(0, 5 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#Set up a subplot grid that has height 2 and width 1 and set the first such plot as active
plt.subplot(2, 1, 1)

#Make the first plot
plt.plot(x, y_sin, 'b--')
plt.title('Sine')

#Set the second subplot as active, and make the second plot
plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'r--')
plt.title('Cosine')
plt.subplots_adjust(hspace=0.5) 
plt.show()
```

```python
#Bar plot
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y, align='center')
plt.bar(x2, y2, color='g')
plt.title('Bar Graph')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
```

# Box plot

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data, vert=False, patch_artist=False)
```

# Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Javascript', 'Java'
sizes = [245, 210, 130, 215]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)
#Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
