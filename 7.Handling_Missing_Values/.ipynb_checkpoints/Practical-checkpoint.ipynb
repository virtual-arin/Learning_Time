# Handling missing values

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum()) # Finds null values
```

### To fill missing data, we can impute with the mean, median (numerical), or mode (categorical)

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
filled_df = df
filled_df['Age'] = filled_df['Age'].fillna(df['Age'].mean())
filled_df.info()
```

### Forward & Backward filling

```python
backward_fill = df
backward_fill['Gender'] = backward_fill['Gender'].bfill()
```

```python
backward_fill.info()
```

## Handling missing values
