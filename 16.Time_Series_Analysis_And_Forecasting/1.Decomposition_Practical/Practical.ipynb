```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('AirPassengers.csv')
```

```python
df['Month'] = pd.to_datetime(df['Month'])
```

```python
df.set_index('Month', inplace=True)
```

```python
df.index.freq = 'MS'
```

```python
df.head()
```

```python
# Perform Seasonal Decompose
multiplicative_result = seasonal_decompose(df['#Passengers'], model='multiplicative')
```

```python
# Plot the components
plt.rcParams.update({'figure.figsize': (10, 10)})
multiplicative_result.plot()
plt.show()
```

```python
# Perform Seasonal Decompose
additive_result = seasonal_decompose(df['#Passengers'], model='additive')
# Plot the components
plt.rcParams.update({'figure.figsize': (10, 10)})
additive_result.plot()
plt.show()
```
