```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Churn_Modelling.csv')
```

```python
data.head()
```

# Univariate Analysis

```python
new_data = data[['Geography', 'Gender', 'Exited']]
```

```python
new_data.Exited.value_counts()
```

```python
new_data.Exited.value_counts()/len(new_data)*100 # into percentage
```

```python
new_data.head()
```

```python
for i, predictor in enumerate(new_data.drop(columns=['Exited'])):
    plt.figure()
    sns.countplot(data=new_data, x=predictor, hue='Exited')
```

# Bivariate Analysis

```python
sns.histplot(x='Gender', hue='Geography', data=new_data, stat='count', multiple='dodge')
```

```python
alternate_new_data = new_data.loc[new_data['Exited']==1]
```

```python
sns.histplot(x='Gender', hue='Geography', data=alternate_new_data, stat='count', multiple='dodge')
```

# Numerical Analysis

```python
data.corr(numeric_only=True)
```

```python
plt.figure(figsize=(20,8))
data.corr(numeric_only=True)['Exited'].sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(data.corr(numeric_only=True), cmap='YlGnBu')
```

```python
data['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
new_data2 = data.loc[data['Exited']==1]
```

```python
new_data2['Age'].value_counts().sort_index(ascending=True).plot()
```

```python
Tot = sns.kdeplot(data.Age[(data["Exited"]==0) ],
                color="Red", shade=True)
Tot = sns.kdeplot(data.Age[(data["Exited"]==1) ],
                color="Blue", shade=True)
Tot.legend(["No Churn", "Churn"], loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Age')
Tot.set_title('Age by churn')
```
