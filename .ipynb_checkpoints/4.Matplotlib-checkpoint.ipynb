# Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

* Create publication quality plots.
* Make interactive figures that can zoom, pan, update.
* Customize visual style and layout.
* Export to many file formats.
* Embed in JupyterLab and Graphical User Interfaces.
* Use a rich array of third-party packages built on Matplotlib.

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
#plotting
plt.scatter(x,y, c='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D Graph')
plt.savefig('sample.png')
```

```python
y = x * x
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y,'r*-')
```

```python
plt.plot(x,y,'ro')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed', linewidth=2, markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2D-Graph')
```

```python
#creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'b--')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.title('Matplotlib graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y)
plt.show()
```
