# Operators:
* Operators are special symbols or keywords that perform operations on one or more values or variables. They represent computations or actions to be performed. Python offers various types of operators, including:
* Arithmetic Operators: For mathematical calculations (e.g., +, -, *, /, **, %, //).
* Assignment Operators: For assigning values to variables (e.g., =, +=, -=, *=).
* Comparison Operators: For comparing values and returning Boolean results (e.g., ==, !=, >, <, >=, <=).
* Logical Operators: For combining or modifying Boolean conditions (e.g., and, or, not).
* Identity Operators: For checking if two variables refer to the same object in memory (e.g., is, is not).
* Membership Operators: For checking if a value exists within a sequence (e.g., in, not in).

# Operands:

# Sets in python
* In Python, a set is an unordered collection of unique elements. It is a mutable data type, meaning you can add or remove elements after creation. Sets are particularly useful for tasks involving membership testing, removing duplicates from sequences, and performing mathematical set operations like union, intersection, and difference. 

```python
friends = {'Advik', 'Arin', 'Vishal', 'Vikas', 'Advik', 'Arin'}
print(friends)
```

```python
type(friends)
```

```python
len(friends)
```

```python
friends.add('Rishita')
print(friends)
```

```python
friends.remove('Vikas')
print(friends)
```

```python
# Union Set: unites unique elements
Boys = {'Advik', 'Arin', 'Vishal'}
Girls = {'Vaishnavi', 'Rishita', 'Radhika'}
print(Boys | Girls)
```

```python
print(Boys.union(Girls))
```

```python
print(len(Boys.union(Girls)))
```

```python
# Intersection Set: outputs common elements
multiple_of_2 = {2, 4, 6, 8, 10, 12}
multiple_of_4 = {4, 8, 12, 16, 20}
print(multiple_of_2 & multiple_of_4)
```

```python
print(multiple_of_2.intersection(multiple_of_4))
```

```python
# Difference: difference between two sets
print(multiple_of_2 - multiple_of_4)
```

```python
print(multiple_of_2.difference(multiple_of_4))
```

```python
# Symmetric difference: results in a new set containing all elements that are present in either set A or set B
print(multiple_of_2 ^ multiple_of_4)
```

```python
print(multiple_of_2.symmetric_difference(multiple_of_4))
```

# Functions in python
* In Python, a function is a block of organized, reusable code that performs a specific task. Functions are defined using the def keyword.
* Lambda functions in Python are small, anonymous functions defined using the lambda keyword. They are also known as anonymous functions because they do not require a name, unlike functions defined with def.

```python
num_sum = lambda x: x + 2
print(num_sum(21))
```

```python
def even_or_odd(num):
    if num % 2  == 0:
        print("Even number")
    else:
        print("Odd number")

even_or_odd(15)
even_or_odd(22)
```

```python
def factorial(num):
    fact = 1
    for i in range(1, num + 1):
        fact = fact * i
    return fact
factorial(5)
```

```python
def natural_num_sum(num):
    n = 0
    for i in range(1, num + 1):
        n = n + i
    return n
natural_num_sum(5)
```

```python
def super_hero(name, power):
    print("Superhero Name:", name)
    print("Superhero Power:", power)
super_hero("Ironman", 1000)
```

```python
def super_hero(name, power=500):
    print("Superhero Name:", name)
    print("Superhero Power:", power)
super_hero("Hulk")
```

```python
# Lambda function that adds 10 to a number
addition = lambda x: x + 10
addition(20)
```

```python
# Lambda function that adds two numbers
addition = lambda x, y: x + y
addition(20, 20)
```

```python
# Check odd or even
odd_or_even = lambda num: "Even" if num % 2 == 0 else "Odd"
odd_or_even(22)
```
