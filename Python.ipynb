# Datatypes in python
* int: Whole numbers, positive or negative, without a decimal component.
* float: Numbers with a decimal component.
* complex: Numbers with a real and an imaginary part (e.g., a + bj).
* str: Sequences of characters enclosed in single or double quotes. Python does not have a separate character type; a single character is a string of length one.
* list: Ordered, mutable collections of items. Items can be of different data types. Defined using square brackets [].
* tuple: Ordered, immutable collections of items. Items can be of different data types. Defined using parentheses ().
* range: Immutable sequence of numbers, often used in loops.
* dict: Unordered collections of key-value pairs. Keys must be unique and immutable. Defined using curly braces {}.
* set: Unordered collections of unique, immutable items. Defined using curly braces {} or the set() constructor.
* frozenset: Immutable versions of sets.
* bool: Represents truth values, either True or False.
* NoneType: Represents the absence of a value.

# Operators:
* Operators are special symbols or keywords that perform operations on one or more values or variables. They represent computations or actions to be performed. Python offers various types of operators, including:
* Arithmetic Operators: For mathematical calculations (e.g., +, -, *, /, **, %, //).
* Assignment Operators: For assigning values to variables (e.g., =, +=, -=, *=).
* Comparison Operators: For comparing values and returning Boolean results (e.g., ==, !=, >, <, >=, <=).
* Logical Operators: For combining or modifying Boolean conditions (e.g., and, or, not).
* Identity Operators: For checking if two variables refer to the same object in memory (e.g., is, is not).
* Membership Operators: For checking if a value exists within a sequence (e.g., in, not in).

# Operands:
* Operands are the values, variables, or expressions on which an operator performs its operation. They are the "inputs" to the operator.

# Tuples in python
### Tuples:
* Ordered and unchangeable (immutable) collections of items. They are defined using parentheses ().
* Tuples are immutable, meaning their elements cannot be changed, added, or removed after the tuple is created.
* Ideal for storing collections of data that should remain constant, such as geographic coordinates, database records, or configuration settings.
