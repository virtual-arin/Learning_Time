```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

# Standardization (Z-Score Scaling) 
* What it does: Centers data around zero and scales it by the standard deviation, transforming data to follow a standard normal distribution (mean=0, std dev=1).
* When to use: Algorithms that assume normally distributed data or use distance metrics (SVM, PCA, Linear Regression).
* Outliers: Less affected by outliers as it uses mean and standard deviation.
# Normalization (Min-Max Scaling) 
* What it does: Rescales data to a fixed range, usually [0] or [-1].
* When to use: Algorithms sensitive to feature magnitudes and where the distribution isn't Gaussian (Neural Networks, k-NN).
* Outliers: Highly sensitive to outliers, which can compress the rest of the data.

```python
new_df = pd.DataFrame(df, columns=['Age', 'Tenure'])
```

```python
new_df.head()
```

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
```

```python
scaler = MinMaxScaler() # instantiating min max scaler function
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

```python
scaler = StandardScaler() # instantiating standard scaler function
standardized_df = scaler.fit_transform(new_df)
print(standardized_df)
```
