```python
# Target Encoding
from category_encoders import TargetEncoder
df1 = pd.read_csv('Churn_Modelling.csv')
df1
```

```python
df1.drop(columns=['RowNumber', 'CustomerId', 'Surname'], axis=1, inplace=True)
```

```python
df1['Gender'] = df1['Gender'].fillna('Male')
```

```python
df1.head()
```

```python
encoder = TargetEncoder()
df1['Gender_Encoded'] = encoder.fit_transform(df1['Gender'], df1['Exited'])
```

```python
df1.head()
```

```python
df1['Gender_Encoded'].value_counts()
```

* Target encoding offers an easy benefit in its ability to capture complex, non-linear relationships between categorical features and the target variable, often improving model accuracy. A key limitation is the risk of overfitting the training data, especially with small datasets or high-cardinality features, requiring careful regularization
